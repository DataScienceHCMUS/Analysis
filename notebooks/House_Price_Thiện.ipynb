# Tiền xử lý

### Xóa các cột thiếu quá nhiều

```python
df = df.drop(["Id","PoolQC","MiscFeature","Alley","Fence","FireplaceQu","LotFrontage"],axis=1)
df
```

```python
col = ['MSZoning', 'Street', 'LotShape', 'LandContour', 'Utilities', 'LotConfig','LandSlope','Neighborhood','Condition1','Condition2','BldgType','HouseStyle',
                'RoofStyle', 'RoofMatl', 'Exterior1st','Exterior2nd','MasVnrType','ExterQual','ExterCond','Foundation','BsmtQual', 'BsmtCond','BsmtExposure','BsmtFinType1','BsmtFinType2',
                 'Heating', 'HeatingQC','CentralAir','Electrical','KitchenQual','Functional','GarageType','GarageFinish','GarageQual','GarageCond','PavedDrive',
                'SaleType','SaleCondition']

df = df.drop(columns=col,axis = 1)
```

### Điền dữ liệu bị thiếu

```python
from sklearn.impute import SimpleImputer
import numpy as np
imp_2 = SimpleImputer(missing_values=np.nan, strategy='median')
train = imp_2.fit_transform(df)
train = pd.DataFrame(train)
train.columns = df.columns # This line adds the column names back to the corresponding features
train
```

### Trực quan thống kê của các cột

```python
import matplotlib.pyplot as plt
import seaborn as sns
df1 = pd.melt(df, value_vars=df.columns, id_vars=['SalePrice'])
g = sns.FacetGrid(df1, col='variable', col_wrap=3, height=3, aspect=1.5, sharex=False, sharey=False)
g.map(sns.distplot, "value")
```

### Trực quan mối tương quan giữa các cột

```python
plt.figure(figsize=(32, 32))
corr = df.corr()
sns.heatmap(data=corr, annot=True, cmap='Blues_r')
```

```python
cor_col = ['OverallQual', 'GrLivArea', 'GarageCars', 'TotalBsmtSF', 'FullBath', 'TotRmsAbvGrd', 'YearBuilt', 'YearRemodAdd']

df2 = pd.melt(df, value_vars=cor_col, id_vars=['SalePrice'])
g = sns.FacetGrid(df2, col='variable', col_wrap=3, height=3, aspect=1.5, sharex=False, sharey=False)
g.map(sns.regplot, "value", "SalePrice", color='purple', scatter_kws={'alpha':0.4}, line_kws={'color':'red'})
```
